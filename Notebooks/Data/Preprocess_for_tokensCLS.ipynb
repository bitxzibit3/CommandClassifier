```python
import json
from collections import defaultdict
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

# Сформировать множество из разных файлов

```python
# df = pd.read_csv("../../Data/Raw/one_cmd_v3_03062022.csv", sep=";")
data = []
with open("../../Data/Raw/data_v5.jsonl", "r") as f:
    for line in f:
        data.append(json.loads(line))
```

```python
labels = []
for i in range(len(data[0][2])):
    labels.append([x[2][i] for x in data])
```

```python
counter = 0
for sample in data:
    for a_i, attr_label in enumerate(["D", "M", "DH", "O1", "N", "R1", "O2", "R2", "O3", "S", "G"]):
        if sample[2][a_i+1]>0:
            if attr_label not in sample[1]:
                counter += 1
                print(counter, sample)
```

```python
counter = 0
for sample in data:
    if "A" not in sample[1]:
        counter += 1
        print(counter, sample)
```

# сплит

```python
# сплит сгенерированной части на трейн и валид 
validIndex = set()
for i in range(len(data[0][2])):
    train_x, valid_x = train_test_split(list(range(len(data))), test_size=0.03, random_state=69, shuffle=True, stratify=labels[i])
    validIndex |= set(valid_x)
len(validIndex) / len(data)
```

```python
for i in range(len(data)):
    if i in validIndex:
        data[i].append("valid")
    else:
        data[i].append("train")
```

```python
with open("../../Data/Interim/data_v5_split.jsonl", "w") as f:
    for d in data:
        f.write(json.dumps(d)+"\n")
```
